```python
import statsmodels.api as sm

d2['intercept'] = 1

X = d2[['intercept', 'kast']]

y = d2['Win_loss']

logit_model = sm.Logit(y, X)
result = logit_model.fit()

print(result.summary())
odds_ratios = result.params
print("\nOdds Ratios:")
print(odds_ratios)

y_prob = result.predict(X)


df_predictions = pd.DataFrame({'Actual': y, 'Predicted Probability': y_prob})


df_predictions = df_predictions.sort_values(by='Predicted Probability')


plt.figure(figsize=(10, 6))
sns.lineplot(x=range(len(df_predictions)), y=df_predictions['Predicted Probability'], label='Predicted Probability')
sns.scatterplot(x=range(len(df_predictions)), y=df_predictions['Actual'], color='red', label='Actual Outcome', s=10)
plt.title('Binary Logistic Regression: Predicted Probabilities vs. Actual Outcomes')
plt.xlabel('Observations')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
